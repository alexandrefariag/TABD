```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
import pyspark.sql.functions as f
from pyspark.sql.types import *

spark = SparkSession.builder.master('local[2]').getOrCreate()
```

```python


globalmobilepenetration = spark.read.csv("hdfs://hdfs-nn:9000/user/Projeto TABD/Bronze/Global-mobile-social-network-penetration.csv", 
                                                 header=False, 
                                                 sep=";")
```

```python
globalmobilepenetration.printSchema()
globalmobilepenetration.show()
globalmobilepenetration.toPandas()
```

```python
globalmobilepenetration = globalmobilepenetration.selectExpr("`_c0` as region",
                                                             "'0' as social_media_users",
                                                             "'2020' as year",
                                                             "`_c1` as mobile_penetration_percentage",
                                                            "'0' as social_media_users_percentage")
## Trocar a ordem para ficar igual aos outros
globalmobilepenetration = globalmobilepenetration[['region', 'social_media_users', 'year', 'mobile_penetration_percentage', 'social_media_users_percentage']]


## Converter o tipo da coluna adicionada para ficar igual

globalmobilepenetration = globalmobilepenetration.withColumn(
    "social_media_users",
    globalmobilepenetration["social_media_users"].cast(DoubleType()))
globalmobilepenetration = globalmobilepenetration.withColumn(
    "year",
    globalmobilepenetration["year"].cast(IntegerType()))

globalmobilepenetration = globalmobilepenetration.withColumn(
    "mobile_penetration_percentage",
    globalmobilepenetration["mobile_penetration_percentage"].cast(IntegerType()))

globalmobilepenetration = globalmobilepenetration.withColumn(
    "social_media_users_percentage",
    globalmobilepenetration["social_media_users_percentage"].cast(FloatType()))

globalmobilepenetration.printSchema()
globalmobilepenetration.toPandas()
```

```python
globalmobilepenetration = globalmobilepenetration.withColumn(
    "mobile_penetration_percentage",
    globalmobilepenetration["mobile_penetration_percentage"].cast(IntegerType()))
globalmobilepenetration.printSchema()
```

```python
globalmobilepenetration \
    .write \
    .format("parquet") \
    .mode("overwrite") \
    .save("hdfs://hdfs-nn:9000/warehouse/tabd.db/GlobalMobilePenetration/")
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
import pyspark.sql.functions as f
from pyspark.sql.types import *
from pyspark.ml import *

spark = SparkSession.builder.master('local[2]').getOrCreate()
```

```python


numberofworldwideinternetusersbyregion = spark.read.csv("hdfs://hdfs-nn:9000/user/Projeto TABD/Bronze/NumberOfWorldwideInternetUsersByRegion.csv", 
                                                 header=True, 
                                                 sep=";")
```

```python
numberofworldwideinternetusersbyregion.printSchema()
numberofworldwideinternetusersbyregion.show()
numberofworldwideinternetusersbyregion.toPandas()
```

```python
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.selectExpr("`Number of worldwide internet users 2009-2020, by region` as year", 
                                                                             "_c1 as Asia", 
                                                                             "_c2 as Europe",
                                                                             "_c3 as `North America`", 
                                                                             "_c4 as `Latin America / Caribbean`",
                                                                             "_c5 as Africa",
                                                                             "_c6 as `Middle East`",
                                                                             "_c7 as `Oceania / Australia`",
                                                                            )
```

```python
numberofworldwideinternetusersbyregion.toPandas()
```

```python

# Substitui o valor  por None para depois remover a linha toda
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "year",
    when(
        (col("year") == "String"), 
        None
    ).otherwise(concat(lit(""), col("year")))
)
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "Asia",
    when(
        (col("Asia") == "String"), 
        None
    ).otherwise(concat(lit(""), col("Asia")))
)
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "Europe",
    when(
        (col("Europe") == "String"), 
        None
    ).otherwise(concat(lit(""), col("Europe")))
)
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "North America",
    when(
        (col("North America") == "String"), 
        None
    ).otherwise(concat(lit(""), col("North America")))
)
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "Latin America / Caribbean",
    when(
        (col("Latin America / Caribbean") == "String"), 
        None
    ).otherwise(concat(lit(""), col("Latin America / Caribbean")))
)
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "Africa",
    when(
        (col("Africa") == "String"), 
        None
    ).otherwise(concat(lit(""), col("Africa")))

)
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "Middle East",
    when(
        (col("Middle East") == "String"), 
        None
    ).otherwise(concat(lit(""), col("Middle East")))
)
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "Oceania / Australia",
    when(
        (col("Oceania / Australia") == "String"), 
        None
    ).otherwise(concat(lit(""), col("Oceania / Australia")))

)
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.dropna(how='any')
```

```python
numberofworldwideinternetusersbyregion.toPandas()
```

```python
# Unpivot - Columns to rows
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.select(
        col("year"),
        expr("stack(7, 'Asia', `Asia`, 'Europe', `Europe`, 'North America', `North America`, 'Latin America / Caribbean', `Latin America / Caribbean`, 'Africa', `Africa`, 'Middle East', `Middle East`, 'Oceania / Australia', `Oceania / Australia`) as (region,internet_users)")
)
```

```python
numberofworldwideinternetusersbyregion.toPandas()
```

```python
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "internet_users",
    f.regexp_replace(col("internet_users"),",", "."))

numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "internet_users",
    f.regexp_replace(col("internet_users")," ", ""))


numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "year",
    numberofworldwideinternetusersbyregion["year"].cast(IntegerType()))

numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "internet_users",
    numberofworldwideinternetusersbyregion["internet_users"].cast(DoubleType()))

## Inserir o resto das colunas para ficar igual ao resto das tabelas 

numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.selectExpr("year as year",
    "region as region", 
                                                                                           "internet_users as internet_users",
                                                                             "'null' as individuals",
                                                                                           
                                                                             "'0' as activity_percentage", 
                                                                             "'0' as usage_rate",
                                                                                           "'null' as gender",
                                                                                           "'0' as penetration_percentage",
                                                                                           
                                                                                           "'null' as internet_activity",
                                                                                           "'null' as frequency_of_access",
                                                                                           "'0' as frequency_of_access_percentage"
                                                                            )
## Trocar a ordem para ficar igual aos outros
numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion[['year', 'region', 'gender', 'usage_rate', 'penetration_percentage', 'internet_users', 'internet_activity', 'activity_percentage', 'frequency_of_access', 'individuals', 'frequency_of_access_percentage']]


## Converter o tipo da coluna adicionada para ficar igual

numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "usage_rate",
    numberofworldwideinternetusersbyregion["usage_rate"].cast(DoubleType()))

numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "penetration_percentage",
    numberofworldwideinternetusersbyregion["penetration_percentage"].cast(IntegerType()))


numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "activity_percentage",
    numberofworldwideinternetusersbyregion["activity_percentage"].cast(IntegerType()))

numberofworldwideinternetusersbyregion = numberofworldwideinternetusersbyregion.withColumn(
    "frequency_of_access_percentage",
    numberofworldwideinternetusersbyregion["frequency_of_access_percentage"].cast(IntegerType()))
```

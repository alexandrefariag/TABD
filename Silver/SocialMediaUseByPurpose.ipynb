```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
import pyspark.sql.functions as f
from pyspark.sql.types import *

spark = SparkSession.builder.master('local[2]').getOrCreate()
```

```python


socialmediausebypurpose = spark.read.csv("hdfs://hdfs-nn:9000/user/Projeto TABD/Bronze/SocialMediaUseByPurpose.csv", 
                                                 header=True, 
                                                 sep=";")
```

```python
socialmediausebypurpose.printSchema()
socialmediausebypurpose.show()
socialmediausebypurpose.toPandas()
```

```python
socialmediausebypurpose = socialmediausebypurpose.selectExpr("`Data extracted on 02/04/2021 16:43:43 from [ESTAT]` as region", 
                                                            "_c1 as purpose", 
                                                            "_c2 as a2013",
                                                            "_c3 as Remove",
                                                            "_c4 as a2015",
                                                            "_c5 as Remove",
                                                            "_c6 as a2017",
                                                            "_c7 as Remove",
                                                            "_c8 as a2019",
                                                            "_c9 as Remove")
```

```python
socialmediausebypurpose.toPandas()
```

```python
socialmediausebypurpose = socialmediausebypurpose.filter((socialmediausebypurpose.purpose != 'Social media use by purpose [ISOC_CISMP__custom_767933]'))

socialmediausebypurpose = socialmediausebypurpose.filter((socialmediausebypurpose.purpose != '11/03/2021 11:00'))

socialmediausebypurpose = socialmediausebypurpose.filter((socialmediausebypurpose.purpose != 'TIME'))

socialmediausebypurpose = socialmediausebypurpose.filter((socialmediausebypurpose.purpose != 'INDIC_IS (Labels)'))

socialmediausebypurpose = socialmediausebypurpose.filter((socialmediausebypurpose.purpose != 'low reliability'))

socialmediausebypurpose = socialmediausebypurpose.filter((socialmediausebypurpose.purpose != 'not significant'))

socialmediausebypurpose = socialmediausebypurpose.filter((socialmediausebypurpose.purpose != 'not available'))
```

```python
socialmediausebypurpose.toPandas()
```

```python
socialmediausebypurpose = socialmediausebypurpose.drop("Remove")
```

```python
socialmediausebypurpose.toPandas()
```

```python
# Unpivot - Columns to rows
socialmediausebypurpose = socialmediausebypurpose.select(
        col("region"), col("purpose"),
        expr("stack(4, 'a2013', `a2013`, 'a2015', `a2015`, 'a2017', `a2017`, 'a2019', `a2019`) as (year,percentage)")
)
```

```python
socialmediausebypurpose.toPandas()
```

```python
socialmediausebypurpose = socialmediausebypurpose.withColumn(
    "percentage",
    f.regexp_replace(col("percentage"),":", "0"))

socialmediausebypurpose = socialmediausebypurpose.withColumn(
    "year",
    f.regexp_replace(col("year"),"a", ""))

socialmediausebypurpose = socialmediausebypurpose.withColumn(
    "year",
    socialmediausebypurpose["year"].cast(IntegerType()))

socialmediausebypurpose = socialmediausebypurpose.withColumn(
    "percentage",
    socialmediausebypurpose["percentage"].cast(IntegerType()))

socialmediausebypurpose = socialmediausebypurpose.withColumn(
    "region",
    when(
        (col("region") == "European Union - 27 countries (from 2020)"), "European Union from 2020"
    ).when(
        (col("region") == "European Union - 28 countries (2013-2020)"), "European Union from 2013 to 2020"
    ).when(
        (col("region") == "European Union - 27 countries (2007-2013)"), "European Union from 2007 to 2013"
    ).when(
        (col("region") == "European Union - 15 countries (1995-2004)"), "European Union from 1995 to 2004"
    ).when(
        (col("region") == "Germany (until 1990 former territory of the FRG)"), "Germany" 
    ).when(
        (col("region") == "Euro area (EA11-1999, EA12-2001, EA13-2007, EA15-2008, EA16-2009, EA17-2011, EA18-2014, EA19-2015)"), "Euro area"
    ).otherwise(col("region")))
```

```python
socialmediausebypurpose.toPandas()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.functions import *
import pyspark.sql.functions as f
from pyspark.sql.types import *

spark = SparkSession.builder.master('local[2]').getOrCreate()
```

```python
data_file="hdfs://hdfs-nn:9000/user/Projeto TABD/Bronze/ListOfCountriesByNumberOfInternetUsers.csv"
listofcountriesbynumberofinternetusers = spark.read.csv(data_file,header=True, sep=",")
```

```python
listofcountriesbynumberofinternetusers.printSchema()
listofcountriesbynumberofinternetusers.show()
listofcountriesbynumberofinternetusers.toPandas()
```

```python
listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.selectExpr("`Country or Area` as country_or_area", 
                                                                                           "`Internet Users` as internet_users",
                                                                                           "Population as population",
                                                                                           "Rank3 as population_rank",
                                                                                           "Percentage as internet_users_percentage",
                                                                                           "Rank5 as internet_user_rank", "'' as country_id", "'0' as status", "'9999-01-01' as join_date", "'0' as area_km", "'0' as current_account_balance", "'0' as internet_hosts", "'0' as `telephone_main_lines_in_use`", "'0' as telephone_mobile_celular"
                                                                                          )

#numberofworldwidesocialmediause = numberofworldwidesocialmediause.selectExpr("Region", "cast(NumberSocialMediaUsers as int) NumberSocialMediaUsers")
```

```python
listofcountriesbynumberofinternetusers.toPandas()
```

```python
listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn(
    "year",
    when(
        (col("internet_users_percentage") == "37.60% (2012)") | 
        (col("internet_users_percentage") == "8.95% (2012)") |
        (col("internet_users_percentage") == "41.03% (2012)"),
        2012
    ).when(
        (col("internet_users_percentage") == "54.55% (2013)") |
        (col("internet_users_percentage") == "86.90% (2013)"),
        2013
    ).when(
        (col("internet_users_percentage") == "94.44% (2016)") |
        (col("internet_users_percentage") == "81.57% (2016)") |
        (col("internet_users_percentage") == "99.02% (2016)"),
        2016
    ).otherwise(2018)
)


listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn(
    "internet_users_percentage",
    when(
        (col("internet_users_percentage") == "41.03% (2012)"), "41.03%"
    ).when(
        (col("internet_users_percentage") == "37.60% (2012)"), "37.60%"
    ).when(
        (col("internet_users_percentage") == "8.95% (2012)"), "8.95%"
    ).when(
        (col("internet_users_percentage") == "54.55% (2013)"), "54.55%"
    ).when(
        (col("internet_users_percentage") == "86.90% (2013)"), "86.90%"
    ).when(
        (col("internet_users_percentage") == "94.44% (2016)"), "94.44%"
    ).when(
        (col("internet_users_percentage") == "81.57% (2016)"), "81.57%"
    ).when(
        (col("internet_users_percentage") == "99.02% (2016)"), "99.02%"
    ).otherwise(col("internet_users_percentage")))

```

```python
listofcountriesbynumberofinternetusers.toPandas()
```

```python

#Remover espaços - é preciso para converter o tipo, se tiver espaços nao converte direito
listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn(
    "internet_users",
    f.regexp_replace(col("internet_users")," ", ""))

listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn(
    "population",
    f.regexp_replace(col("population")," ", ""))

listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn(
    "internet_users_percentage",
    f.regexp_replace(col("internet_users_percentage"),"%", ""))



#Converter tipo
listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn(
    "population",
    listofcountriesbynumberofinternetusers["population"].cast(IntegerType()))
listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn(
    "population_rank",
    listofcountriesbynumberofinternetusers["population_rank"].cast(IntegerType()))

listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn(
    "internet_user_rank",
    listofcountriesbynumberofinternetusers["internet_user_rank"].cast(IntegerType()))

listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn(
    "internet_users",
    listofcountriesbynumberofinternetusers["internet_users"].cast(IntegerType()))

listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn(
    "internet_users_percentage",
    listofcountriesbynumberofinternetusers["internet_users_percentage"].cast(FloatType()))


listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn("current_account_balance", listofcountriesbynumberofinternetusers["current_account_balance"].cast(IntegerType()))
listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn("internet_hosts", listofcountriesbynumberofinternetusers["internet_hosts"].cast(IntegerType()))
listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn("telephone_main_lines_in_use", listofcountriesbynumberofinternetusers["telephone_main_lines_in_use"].cast(IntegerType()))
listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn("telephone_mobile_celular", listofcountriesbynumberofinternetusers["telephone_mobile_celular"].cast(IntegerType()))
listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn("status", listofcountriesbynumberofinternetusers["status"].cast(IntegerType()))
listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers.withColumn("join_date", listofcountriesbynumberofinternetusers["join_date"].cast(DateType()))

listofcountriesbynumberofinternetusers = listofcountriesbynumberofinternetusers[['country_or_area','area_km', 'current_account_balance', 'internet_hosts', 'telephone_main_lines_in_use', 'telephone_mobile_celular', 'country_id', 'status', 'join_date','population', 'population_rank', 'internet_users', 'internet_users_percentage', 'internet_user_rank']]
```

```python
listofcountriesbynumberofinternetusers.printSchema()
```

```python
listofcountriesbynumberofinternetusers \
    .write \
    .format("parquet") \
    .mode("overwrite") \
    .save("hdfs://hdfs-nn:9000/warehouse/tabd.db/ListOfCountriesByNumberOfInternetUsers/")
```

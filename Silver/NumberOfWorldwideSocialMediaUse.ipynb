```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
import pyspark.sql.functions as f
from pyspark.sql.types import *

spark = SparkSession.builder.master('local[2]').getOrCreate()
```

```python


numberofworldwidesocialmediause = spark.read.csv("hdfs://hdfs-nn:9000/user/Projeto TABD/Bronze/NumberOfWorldwideSocialMediaUsers2020ByRegion.csv", 
                                                 header=True, 
                                                 sep=";")
```

```python
numberofworldwidesocialmediause.printSchema()
numberofworldwidesocialmediause.show()
numberofworldwidesocialmediause.toPandas()
```

```python
numberofworldwidesocialmediause.toPandas()
```

```python
numberofworldwidesocialmediause = numberofworldwidesocialmediause.selectExpr("Region as region", 
                                                                             "`Number of worldwide social media users 2020` as social_media_users", 
                                                                             "'2020' as year",
                                                                            "'0' as mobile_penetration_percentage",
                                                                            "'0' as social_media_users_percentage")
numberofworldwidesocialmediause.printSchema()
numberofworldwidesocialmediause.toPandas()
```

```python
numberofworldwidesocialmediause = numberofworldwidesocialmediause.withColumn(
    "social_media_users",
    f.regexp_replace(col("social_media_users")," ", ""))
numberofworldwidesocialmediause = numberofworldwidesocialmediause.withColumn(
    "social_media_users",
    f.regexp_replace(col("social_media_users"),",", "."))

numberofworldwidesocialmediause = numberofworldwidesocialmediause.withColumn(
    "social_media_users",
    numberofworldwidesocialmediause["social_media_users"].cast(DoubleType()))

## Converter o tipo da coluna adicionada para ficar igual
numberofworldwidesocialmediause = numberofworldwidesocialmediause.withColumn(
    "year",
    numberofworldwidesocialmediause["year"].cast(IntegerType()))

numberofworldwidesocialmediause = numberofworldwidesocialmediause.withColumn(
    "mobile_penetration_percentage",
    numberofworldwidesocialmediause["mobile_penetration_percentage"].cast(IntegerType()))

numberofworldwidesocialmediause = numberofworldwidesocialmediause.withColumn(
    "social_media_users_percentage",
    numberofworldwidesocialmediause["social_media_users_percentage"].cast(FloatType()))

numberofworldwidesocialmediause.printSchema()
numberofworldwidesocialmediause.toPandas()

```

```python

numberofworldwidesocialmediause.toPandas()

numberofworldwidesocialmediause \
    .write \
    .format("parquet") \
    .mode("overwrite") \
    .save("hdfs://hdfs-nn:9000/warehouse/tabd.db/NumberOfWorldwideSocialMediaUse/")
```

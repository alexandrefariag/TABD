```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
import pyspark.sql.functions as f
from pyspark.sql.types import *

spark = SparkSession.builder.master('local[2]').getOrCreate()
```

```python
individualsactivities = spark.read.csv("hdfs://hdfs-nn:9000/user/Projeto TABD/Bronze/IndividualsActivities1.csv", header=True, sep=";")
```

```python
individualsactivities.printSchema()
individualsactivities.show()
individualsactivities.toPandas()
```

```python
individualsactivities.toPandas()
```

```python

# Substitui o valor  por None para depois remover a linha toda
individualsactivities = individualsactivities.withColumn(
    "TIME0",
    when(
        (col("TIME0") == "String"), 
        None
    ).otherwise(concat(lit(""), col("TIME0")))
)
individualsactivities = individualsactivities.withColumn(
    "TIME1",
    when(
        (col("TIME1") == "String"), 
        None
    ).otherwise(concat(lit(""), col("TIME1")))
)
individualsactivities = individualsactivities.withColumn(
    "2011",
    when(
        (col("2011") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2011")))
)
individualsactivities = individualsactivities.withColumn(
    "2012",
    when(
        (col("2012") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2012")))
)
individualsactivities = individualsactivities.withColumn(
    "2013",
    when(
        (col("2013") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2013")))
)
individualsactivities = individualsactivities.withColumn(
    "2014",
    when(
        (col("2014") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2014")))
)
individualsactivities = individualsactivities.withColumn(
    "2015",
    when(
        (col("2015") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2015")))
)
individualsactivities = individualsactivities.withColumn(
    "2016",
    when(
        (col("2016") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2016")))
)
individualsactivities = individualsactivities.withColumn(
    "2017",
    when(
        (col("2017") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2017")))
)
individualsactivities = individualsactivities.withColumn(
    "2018",
    when(
        (col("2018") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2018")))
)
individualsactivities = individualsactivities.withColumn(
    "2019",
    when(
        (col("2019") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2019")))
)
individualsactivities = individualsactivities.withColumn(
    "2020",
    when(
        (col("2020") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2020")))
)

individualsactivities = individualsactivities.dropna(how='any')
```

```python
individualsactivities.toPandas()
```

```python
# Unpivot - Columns to rows
individualsactivities = individualsactivities.select(
        col("TIME0"), col("TIME1"),
        expr("stack(10, '2011', `2011`, '2012', `2012`, '2013', `2013`, '2014', `2014`, '2015', `2015`, '2016', `2016`, '2017', `2017`, '2018', `2018`, '2019', `2019`, '2020', `2020`) as (year,activity_percentage)")
)
```

```python
individualsactivities.toPandas()
```

```python

individualsactivities = individualsactivities.selectExpr("`TIME0` as region", 
                                                         "`TIME1` as individuals", 
                                                         "year as year",
                                                         "activity_percentage as activity_percentage"
                                                         )
```

```python
individualsactivities.toPandas()
```

```python

individualsactivities = individualsactivities.withColumn(
    "region",
    when(
        (col("region") == "Euro area (EA11-1999, EA12-2001, EA13-2007, EA15-2008, EA16-2009, EA17-2011, EA18-2014, EA19-2015)"), "Euro Area"   
    ).when(
        (col("region") == "Germany (until 1990 former territory of the FRG)"), "Germany"
    ).when(
        (col("region") == "Kosovo (under United Nations Security Council Resolution 1244/99)"), "Kosovo"
    ).when(
        (col("region") == "European Union - 27 countries (from 2020)"), "European Union from 2020"
    ).when(
        (col("region") == "European Union - 28 countries (2013-2020)"), "European Union from 2013 to 2020"
    ).when(
        (col("region") == "European Union - 27 countries (2007-2013)"), "European Union from 2007 to 2013"    
    ).when(
        (col("region") == "European Union - 25 countries (2004-2006)"), "European Union from 2004 to 2006" 
    ).when(
        (col("region") == "European Union - 15 countries (1995-2004))"), "European Union from 1995 to 2004"  
    ).otherwise(col("region")))


individualsactivities = individualsactivities.withColumn(
    "individuals",
    when(
        (col("individuals") == "Individuals, 15 years old or less"), "Individuals with 15 years old or less"   
    ).when(
        (col("individuals") == "Individuals, 16 to 24 years old"), "Individuals with 16 to 24 years old"
    ).when(
        (col("individuals") == "Individuals, 25 to 34 years old"), "Individuals with 25 to 34 years old"
    ).when(
        (col("individuals") == "Individuals, 35 to 44 years old"), "Individuals with 35 to 44 years old"
    ).when(
        (col("individuals") == "Individuals, 45 to 54 years old"), "Individuals with 45 to 54 years old"
    ).when(
        (col("individuals") == "Individuals, 55 to 64 years old"), "Individuals with 55 to 64 years old"    
    ).when(
        (col("individuals") == "Individuals, 65 to 74 years old"), "Individuals with 65 to 74 years old" 
    ).when(
        (col("individuals") == "Individuals, 75 years old or more"), "Individuals with 75 years old or more"  
    ).otherwise(col("individuals")))
```

```python
individualsactivities.toPandas()
```

```python
individualsactivities = individualsactivities.withColumn(
    "year",
    individualsactivities["year"].cast(IntegerType()))

individualsactivities = individualsactivities.withColumn(
    "activity_percentage",
    individualsactivities["activity_percentage"].cast(IntegerType()))

individualsactivities = individualsactivities.fillna(0)

## Inserir o resto das colunas para ficar igual ao resto das tabelas 

individualsactivities = individualsactivities.selectExpr("year as year",
                                                           "region as region", 
                                                           "'null' as internet_activity",
                                                           "activity_percentage as activity_percentage", 
                                                           "individuals as individuals"
                                                        )

```

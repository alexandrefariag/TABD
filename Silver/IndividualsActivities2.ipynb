```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
import pyspark.sql.functions as f
from pyspark.sql.types import *

spark = SparkSession.builder.master('local[2]').getOrCreate()
```

```python
individualsactivities2 = spark.read.csv("hdfs://hdfs-nn:9000/user/Projeto TABD/Bronze/IndividualsActivities2.csv", header=True, sep=";")
```

```python
individualsactivities2.printSchema()
individualsactivities2.show()
individualsactivities2.toPandas()
```

```python
# Substitui o valor  por None para depois remover a linha toda
individualsactivities2 = individualsactivities2.withColumn(
    "TIME0",
    when(
        (col("TIME0") == "String"), 
        None
    ).otherwise(concat(lit(""), col("TIME0")))
)
individualsactivities2 = individualsactivities2.withColumn(
    "TIME1",
    when(
        (col("TIME1") == "String"), 
        None
    ).otherwise(concat(lit(""), col("TIME1")))
)
individualsactivities2 = individualsactivities2.withColumn(
    "2011",
    when(
        (col("2011") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2011")))
)
individualsactivities2 = individualsactivities2.withColumn(
    "2012",
    when(
        (col("2012") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2012")))
)
individualsactivities2 = individualsactivities2.withColumn(
    "2013",
    when(
        (col("2013") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2013")))
)
individualsactivities2 = individualsactivities2.withColumn(
    "2014",
    when(
        (col("2014") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2014")))
)
individualsactivities2 = individualsactivities2.withColumn(
    "2015",
    when(
        (col("2015") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2015")))
)
individualsactivities2 = individualsactivities2.withColumn(
    "2016",
    when(
        (col("2016") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2016")))
)
individualsactivities2 = individualsactivities2.withColumn(
    "2017",
    when(
        (col("2017") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2017")))
)
individualsactivities2 = individualsactivities2.withColumn(
    "2018",
    when(
        (col("2018") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2018")))
)
individualsactivities2 = individualsactivities2.withColumn(
    "2019",
    when(
        (col("2019") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2019")))
)
individualsactivities2 = individualsactivities2.withColumn(
    "2020",
    when(
        (col("2020") == "String"), 
        None
    ).otherwise(concat(lit(""), col("2020")))
)

individualsactivities2 = individualsactivities2.dropna(how='any')
```

```python
individualsactivities2.toPandas()
```

```python
# Unpivot - Columns to rows
individualsactivities2 = individualsactivities2.select(
        col("TIME0"), col("TIME1"),
        expr("stack(10, '2011', `2011`, '2012', `2012`, '2013', `2013`, '2014', `2014`, '2015', `2015`, '2016', `2016`, '2017', `2017`, '2018', `2018`, '2019', `2019`, '2020', `2020`) as (year,activity_percentage)")
)
```

```python
individualsactivities2.toPandas()
```

```python

individualsactivities2 = individualsactivities2.selectExpr("`TIME0` as region", 
                                                         "`TIME1` as internet_activity", 
                                                         "year as year",
                                                         "activity_percentage as activity_percentage"
                                                         )
```

```python
individualsactivities2.toPandas()
```

```python
individualsactivities2 = individualsactivities2.withColumn(
    "region",
    when(
        (col("region") == "Euro area (EA11-1999, EA12-2001, EA13-2007, EA15-2008, EA16-2009, EA17-2011, EA18-2014, EA19-2015)"), "Euro Area"   
    ).when(
        (col("region") == "Germany (until 1990 former territory of the FRG)"), "Germany"
    ).when(
        (col("region") == "Kosovo (under United Nations Security Council Resolution 1244/99)"), "Kosovo"
    ).when(
        (col("region") == "European Union - 27 countries (from 2020)"), "European Union from 2020"
    ).when(
        (col("region") == "European Union - 28 countries (2013-2020)"), "European Union from 2013 to 2020"
    ).when(
        (col("region") == "European Union - 27 countries (2007-2013)"), "European Union from 2007 to 2013"    
    ).when(
        (col("region") == "European Union - 25 countries (2004-2006)"), "European Union from 2004 to 2006" 
    ).when(
        (col("region") == "European Union - 15 countries (1995-2004))"), "European Union from 1995 to 2004"  
    ).otherwise(col("region")))


individualsactivities2 = individualsactivities2.withColumn(
    "internet_activity",
    when(
        (col("internet_activity") == "Internet use: looking for information about education, training or course offers"), "Looking for information about education, training or course offers"   
    ).when(
        (col("internet_activity") == "Internet use: sending/receiving e-mails"), "Sending/receiving e-mails"
    ).when(
        (col("internet_activity") == "Internet use: telephoning or video calls"), "Telephoning or video calls"
    ).when(
        (col("internet_activity") == "Internet use: participating in social networks (creating user profile, posting messages or other contributions to facebook, twitter, etc.)"), "Participating in social networks"
    ).when(
        (col("internet_activity") == "Internet use: finding information about goods and services"), "Finding information about goods and services"
    ).when(
        (col("internet_activity") == "Internet use: consulting wikis (to obtain knowledge on any subject)"), "Consulting wikis"    
    ).when(
        (col("internet_activity") == "Internet use: reading online news sites/newspapers/news magazines"), "Reading online news sites/newspapers/news magazines" 
    ).when(
        (col("internet_activity") == "Internet use: playing/downloading games, images, films or music"), "Playing/downloading games, images, films or music"  
    ).when(
        (col("internet_activity") == "Internet use: listening to web radios and/or watching web TV"), "Listening to web radios and/or watching web TV"  
    ).when(
        (col("internet_activity") == "Internet use: seeking health information"), "Seeking health information"  
    ).when(
        (col("internet_activity") == "Internet use: playing/downloading games, listening to music or watching videos (excluding VOD)"), "Playing/downloading games, listening to music or watching videos" 
    ).otherwise(col("internet_activity")))
```

```python
individualsactivities2.toPandas()
```

```python
individualsactivities2 = individualsactivities2.withColumn(
    "year",
    individualsactivities2["year"].cast(IntegerType()))

individualsactivities2 = individualsactivities2.withColumn(
    "activity_percentage",
    individualsactivities2["activity_percentage"].cast(IntegerType()))

individualsactivities2 = individualsactivities2.fillna(0)
```

```python
individualsactivities2.toPandas()
```

```python
individualsactivities2 = individualsactivities2.selectExpr("year as year",
                                                           "region as region", 
                                                           "internet_activity as internet_activity",
                                                           "activity_percentage as activity_percentage", 
                                                           "'null' as individuals"
                                                           )
```

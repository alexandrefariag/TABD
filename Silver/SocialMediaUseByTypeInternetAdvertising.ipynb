```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
import pyspark.sql.functions as f
from pyspark.sql.types import *

spark = SparkSession.builder.master('local[2]').getOrCreate()
```

```python


socialmediausebytypeinternetadvertising = spark.read.csv("hdfs://hdfs-nn:9000/user/Projeto TABD/Bronze/SocialMediaUseByTypeInternetAdvertising.csv", 
                                                 header=True, 
                                                 sep=";")
```

```python
socialmediausebytypeinternetadvertising.printSchema()
socialmediausebytypeinternetadvertising.show()
socialmediausebytypeinternetadvertising.toPandas()
```

```python
socialmediausebytypeinternetadvertising = socialmediausebytypeinternetadvertising.selectExpr("`Data extracted on 02/04/2021 19:00:58 from [ESTAT]` as region", 
                                                                             "_c1 as year",
                                                                             "_c2 as `Pay to advertise on the internet`",
                                                                             "_c3 as `Enterprises using information about visitors behaviour on their websites`",
                                                                             "_c4 as `Have a website and pay to advertise on the internet`",
                                                                             "_c5 as `Have web sales to private consumers (B2C) and pay to advertise on the internet`",                
                                                                             "_c6 as `Use social networks (Facebook, LinkedIn, Xing, Viadeo, Yammer, etc)`",     
                                                                             "_c7 as `Use enterprise blog or microblogs Twitter, Present, etc)`", 
                                                                             "_c8 as `Use multimedia content sharing websites (YouTube, Flickr, Picasa, SlideShare, etc)`",
                                                                             "_c9 as `Use wiki based knowledge sharing tools`",       
                                                                             "_c10 as `Use any social media`",
                                                                             "_c11 as `Use only one type of social media`",
                                                                             "_c12 as `Use two or more social media`",
                                                                             "_c13 as `Have a Website and use any social media`",
                                                                             "_c14 as `Website has online ordering, reservation or booking and at least one of: webacc, webctm, webot or webper and use two or more social media`",
                                                                             "_c15 as `Have a formal policy for using social media and have used two or more social media`",   
                                                                             "_c16 as `Do not use the above mentioned social media or use them only for posting paid adverts`",
                                                                             "_c17 as `Use any social media, only for posting paid adverts`")
```

```python
socialmediausebytypeinternetadvertising.toPandas()
```

```python
socialmediausebytypeinternetadvertising = socialmediausebytypeinternetadvertising.filter((socialmediausebytypeinternetadvertising.year != 'None'))

socialmediausebytypeinternetadvertising = socialmediausebytypeinternetadvertising.filter((socialmediausebytypeinternetadvertising.year != 'Social media use by type, internet advertising [ISOC_CISMT__custom_768447]'))

socialmediausebytypeinternetadvertising = socialmediausebytypeinternetadvertising.filter((socialmediausebytypeinternetadvertising.year != '11/03/2021 11:00'))

socialmediausebytypeinternetadvertising = socialmediausebytypeinternetadvertising.filter((socialmediausebytypeinternetadvertising.year != 'TIME'))
```

```python
socialmediausebytypeinternetadvertising.toPandas()
```

```python
# Unpivot - Columns to rows
socialmediausebytypeinternetadvertising = socialmediausebytypeinternetadvertising.select(
        col("region"), col("year"),
        expr("stack(16, 'Pay to advertise on the internet', `Pay to advertise on the internet`, 'Enterprises using information about visitors behaviour on their websites', `Enterprises using information about visitors behaviour on their websites`, 'Have a website and pay to advertise on the internet', `Have a website and pay to advertise on the internet`, 'Have web sales to private consumers (B2C) and pay to advertise on the internet', `Have web sales to private consumers (B2C) and pay to advertise on the internet`, 'Use social networks (Facebook, LinkedIn, Xing, Viadeo, Yammer, etc)', `Use social networks (Facebook, LinkedIn, Xing, Viadeo, Yammer, etc)`, 'Use enterprise blog or microblogs Twitter, Present, etc)', `Use enterprise blog or microblogs Twitter, Present, etc)`, 'Use multimedia content sharing websites (YouTube, Flickr, Picasa, SlideShare, etc)', `Use multimedia content sharing websites (YouTube, Flickr, Picasa, SlideShare, etc)`, 'Use wiki based knowledge sharing tools', `Use wiki based knowledge sharing tools`, 'Use any social media', `Use any social media`, 'Use only one type of social media', `Use only one type of social media`, 'Use two or more social media', `Use two or more social media`, 'Have a Website and use any social media', `Have a Website and use any social media`, 'Website has online ordering, reservation or booking and at least one of: webacc, webctm, webot or webper and use two or more social media', `Website has online ordering, reservation or booking and at least one of: webacc, webctm, webot or webper and use two or more social media`, 'Have a formal policy for using social media and have used two or more social media', `Have a formal policy for using social media and have used two or more social media`, 'Do not use the above mentioned social media or use them only for posting paid adverts', `Do not use the above mentioned social media or use them only for posting paid adverts`, 'Use any social media, only for posting paid adverts', `Use any social media, only for posting paid adverts`) as (advertising_type,percentage)")
)
```

```python
socialmediausebytypeinternetadvertising.toPandas()
```

```python
socialmediausebytypeinternetadvertising = socialmediausebytypeinternetadvertising.withColumn(
    "percentage",
    f.regexp_replace(col("percentage"),":", "0"))

socialmediausebytypeinternetadvertising = socialmediausebytypeinternetadvertising.withColumn(
    "year",
    socialmediausebytypeinternetadvertising["year"].cast(IntegerType()))

socialmediausebytypeinternetadvertising = socialmediausebytypeinternetadvertising.withColumn(
    "percentage",
    socialmediausebytypeinternetadvertising["percentage"].cast(IntegerType()))

socialmediausebytypeinternetadvertising = socialmediausebytypeinternetadvertising.withColumn(
    "region",
    when(
        (col("region") == "European Union - 27 countries (from 2020)"), "European Union from 2020"
    ).when(
        (col("region") == "European Union - 28 countries (2013-2020)"), "European Union from 2013 to 2020"
    ).when(
        (col("region") == "European Union - 27 countries (2007-2013)"), "European Union from 2007 to 2013"
    ).when(
        (col("region") == "European Union - 15 countries (1995-2004)"), "European Union from 1995 to 2004"
    ).when(
        (col("region") == "Germany (until 1990 former territory of the FRG)"), "Germany"
    ).when(
        (col("region") == "Euro area (EA11-1999, EA12-2001, EA13-2007, EA15-2008, EA16-2009, EA17-2011, EA18-2014, EA19-2015)"), "Euro area"
    ).otherwise(col("region")))
```
